```python
user_input = str(input("Enter your palindrome").lower())

user_space = ""

for p in user_input:
  if p.isalnum():
    user_space += p

user_palindrome = user_space[::-1]

if user_palindrome == user_space:
  print(f"{user_palindrome}")
else:
  print(f"This is not a palindrome")
```
