```python
#Fix 1: Used Chatgpt to check if it needed any fixes and found out that I have to start a with 1.

def climbStairs(self, n: int) -> int:
    # a = number of ways to reach the current step
    # b = number of ways to reach the next step
    # Start both at 1 because:
    # - There is 1 way to climb 0 steps
    # - There is 1 way to climb 1 step
    a, b = 1, 1

    # Repeat n times to build up the Fibonacci-like sequence
    for _ in range(n):
        # Update:
        # - 'a' becomes the old 'b' (shift forward)
        # - 'b' becomes the sum of old 'a' and old 'b'
        # This matches the recurrence: f(n) = f(n-1) + f(n-2)
        a, b = b, a + b

    # 'a' now holds the number of ways to climb n steps
    return a
```

```python
def sortColors(self, nums):
  low, mid, high = 0, 0, len(nums)-1
  while mid <= high:
    if nums[mid] == 0:
      nums[low], nums[mid] = nums[mid], nums[low]
      low += 1
      mid += 1
    elif nums[mid] == 1:
      mid += 1
    else:
      nums[mid], nums[high] = nums[high], nums[mid]
      high -= 1
```
